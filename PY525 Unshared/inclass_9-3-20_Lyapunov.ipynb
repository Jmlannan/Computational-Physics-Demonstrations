```
import matplotlib.pyplot as plt
import math
import numpy as np
from sympy import *

def logmap(num, lam):
  x = [.3]
  for i in range(0,num):
    x.append(1- lam* x[-1]*x[-1])
  return x

def lyn(lam, n):
  x = logmap(n,lam)
  c =0
  for i in range(0,n-1):
    c = c + math.log(abs(2*lam*x[i]))
  return c/n

numsx = []
numsy = []
for i in range(1,100):
  numsy.append(lyn(i/100,1000))
  numsx.append(i/1000)
plt.plot(numsx,numsy)

```
