```
from numpy import sqrt
import sympy as sp
%matplotlib inline
sp.init_printing()

phi = [1,(-1 + sqrt(5))/2]
for i in range(1,30):
  #print(phi[i-1] - phi[i])
  phi.append(phi[i-1] - phi[i])
print(phi[-1])
```

```
phia = [1,(-1 + sqrt(5))/2]
phib = [1,(-1 + sqrt(5))/2]
while abs((phia[-1]-phib[-1]) / phib[-1]) < 0.1:
  phib.append(phib[1] * phib[-1])
  phia.append(phia[-2] - phia[-1])
print(len(phib)-1)
```

```
#cas version, results of two algorithms are exactly the same.
import sympy as sp
%matplotlib inline
sp.init_printing()

end = 30

phia = [1,(-1 + sp.sqrt(5))/2]
phib = [1,(-1 + sp.sqrt(5))/2]

for i in range(1,end):
  expa = sp.simplify(phia[-2] - phia[-1])
  expb = sp.simplify(phib[1] * phib[-1])
  phia.append(expa)
  phib.append(expb)

print( "phia: ", phia[-1], " = " , phia[-1].evalf(50),"\n")
print( "phib: ", phib[-1], " = " , phib[-1].evalf(50),"\n")
```
