Consider the advection equation: 
$\frac{\partial  f} {\partial  t} + \tanh(x/R) \frac{\partial  f}{\partial x} = 0$ with the initial condition $f(t=0,x) = \exp[ -x^2 / (2 R^2) ]$, $R=5$. 

Show solutions at $t \approx 2$ using the explicit scheme (i.e. $\frac{\partial  f} {\partial  t} \approx \frac{f^{n+1}_m - f^{n}_m} {\Delta t} $) and  

1.   Central difference for the spatial derivative 
$$ \frac{\partial  f}{\partial x}  \approx \frac{f_{m+1} - f_{m-1} } {2 \Delta x }$$
2.   Forward difference for the spatial derivative 
$$ \frac{\partial  f}{\partial x}  \approx \frac{f_{m+1} - f_{m} } {\Delta x }$$
3.   Backward difference for the spatial derivative 
$$ \frac{\partial  f}{\partial x}  \approx \frac{f_{m} - f_{m-1} } {\Delta x }$$

```
import numpy as np 
import matplotlib.pyplot as plt

R=5 

N=128 # number of lattice sites in x direction 
x_min,x_max= -10*R, 10*R 
delta_x = (x_max-x_min)/N 

delta_t = delta_x/10 # for stability   

def IC(x,xc,R):
  return np.exp( -(x-xc)**2/(2*R**2) )

def forward_difference(f,i):
  dfdx = (f[i+1]-f[i])/delta_t 
  return dfdx 

def backward_difference(f,i):
  dfdx = (f[i] - f[i-1])/delta_t
  return dfdx 

def central_difference(f,i):
  dfdx = (f[i+1] - f[i-1])/(2*delta_t)
  return dfdx 

def x_of_int(i):
  return x_min + i*delta_x

def stepper(f):
  # forward scheme
  f_n = np.zeros(len(f))
  for i in range(1,len(f)-1):
    arg = x_of_int(i)/R
    f_n[i] = f[i] - delta_t*np.tanh(arg)*forward_difference(f,i)
  return f_n 
```

```
f = np.zeros(N)
for i in range(N):
  f[i] = IC(x_of_int(i),0,R)
plt.plot(f,label='Initial')

t=0
for j in range (int(2/delta_t)+1):
  f=stepper(f)
  t = t + delta_t

plt.plot(f,label='Final')
plt.legend()
```

```
import numpy as np 
import matplotlib.pyplot as plt

R=5 

N=128 # number of lattice sites in x direction 
x_min,x_max= -10*R, 10*R 
delta_x = (x_max-x_min)/N 

delta_t = delta_x/10 # for stability   

def IC(x,xc,R):
  return np.exp( -(x-xc)**2/(2*R**2) )

def forward_difference(f,i):
  dfdx = (f[i+1]-f[i])/delta_t 
  return dfdx 

def backward_difference(f,i):
  dfdx = (f[i] - f[i-1])/delta_t
  return dfdx 

def central_difference(f,i):
  dfdx = (f[i+1] - f[i-1])/(2*delta_t)
  return dfdx 

def x_of_int(i):
  return x_min + i*delta_x

def stepper(f):
  # forward scheme
  f_n = np.zeros(len(f))
  for i in range(1,len(f)-1):
    arg = x_of_int(i)/R
    if np.sign(np.tanh(arg)) > 0:
      f_n[i] = f[i] - delta_t*np.tanh(arg)*backward_difference(f,i)
    else:
      f_n[i] = f[i] - delta_t*np.tanh(arg)*forward_difference(f,i)
  return f_n 

  f = np.zeros(N)
for i in range(N):
  f[i] = IC(x_of_int(i),0,R)
plt.plot(f,label='Initial')

t=0
for j in range (int(2/delta_t)+1):
  f=stepper(f)
  t = t + delta_t

plt.plot(f,label='Final')
plt.legend()

print("peak: ", 1-np.max(f))
```
