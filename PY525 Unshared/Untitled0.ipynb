```
import numpy as np
import math
import matplotlib.pyplot as plt

gs = (np.sqrt(5) -1)/ 2
gs2 = gs**2
eps = 1E-7

#starting range
a = 0 
b = 5

#initial divisions
x1 = a + gs2*(b-a)
x2 = a + gs*(b-a)

while b-a > eps:
  y1 = math.cos(x1) +1
  y2 = math.cos(x2) +1
  if y1 < y2:
    b = x2
    x2= x1
    x1 = a + gs2*(b-a)
  else:
    a = x1
    x1=x2
    x2 = a + gs*(b-a)

print((a+b)/2 - math.pi)
```
