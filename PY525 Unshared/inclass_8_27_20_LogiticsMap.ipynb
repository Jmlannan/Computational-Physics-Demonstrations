Joseph Lannan and Elija Numa, Logistics Map and Bifurcation Diagrams


```
import matplotlib.pyplot as plt

def logmap(num, lam):
  x = [.3]
  for i in range(1,num):
    x.append(1- lam* x[-1]*x[-1])
  return x

plt.plot(logmap(60,.5), "o")
plt.plot(logmap(60,.5))
plt.title("lam = .5")
plt.show

plt.figure()
plt.plot(logmap(60,.8), "o")
plt.plot(logmap(60,.8))
plt.title("lam = .8")
plt.show

plt.figure()
plt.plot(logmap(60,1), "o")
plt.plot(logmap(60,1))
plt.title("lam = 1")
plt.show

plt.figure()
plt.plot(logmap(60,1.3), "o")
plt.plot(logmap(60,1.3))
plt.title("lam = 1.3")
plt.show

plt.figure()
plt.plot(logmap(60,1.5), "o")
plt.plot(logmap(60,1.5))
plt.title("lam = 1.5")
plt.show

plt.figure()
plt.plot(logmap(200,1.76), "o")
plt.plot(logmap(200,1.76))
plt.title("lam = 1.8")
plt.show

plt.figure(figsize= [20,20])
plt.title("Bifurcation Diagram")
plt.xlabel("$\lambda$")
plt.ylabel("Value")
for i in range(0,1000):
  valy = []
  valx = []
  valx.append([i/(1000/2) for n in range(0,100)])
  valy.append(logmap(1100,i/(1000/2))[1000:1100])
  
  plt.plot(valx[-1],valy[-1],"k.")

plt.figure(figsize= [20,20])
plt.title("Bifurcation Diagram")
plt.xlabel("$\lambda$")
plt.ylabel("Value")
for i in range(1750,2000):
  valy = []
  valx = []
  valx.append([i/(1000) for n in range(0,100)])
  valy.append(logmap(1100,i/(1000))[1000:1100])
  
  plt.plot(valx[-1],valy[-1],"k.")
```
